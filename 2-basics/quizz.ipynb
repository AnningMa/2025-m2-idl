## Hard

```python
print(torch.arange(10)[..., None]) #在切片操作中，None 的意思是“在这里插入一个大小为 1 的新维度”。
print(torch.arange(10))
```

```python
torch.abs(torch.arange(10)[..., None] - torch.arange(10))
```

```python
torch.arange(10)[:, None] == torch.arange(10)
```
